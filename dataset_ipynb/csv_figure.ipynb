```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to generate figures from CSV
def generate_figures(csv_path):
    try:
        # Load the CSV file
        df = pd.read_csv(csv_path)
        
        # 1. Display First Few Rows (Text Preview as a Figure)
        fig1, ax = plt.subplots(figsize=(10, 6))
        ax.axis('off')
        table = pd.plotting.table(ax, df.head(10), loc='center', cellLoc='center', colWidths=[0.2]*len(df.columns))
        table.auto_set_font_size(False)
        table.set_fontsize(10)
        table.scale(1.2, 1.2)
        plt.title("Table: First 10 Rows of the Dataset")
        plt.savefig("figure1_table_preview.png")
        plt.show()
        
        # 2. Class Distribution Bar Plot
        if "label" in df.columns:
            plt.figure(figsize=(8, 6))
            df['label'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
            plt.title("Class Distribution")
            plt.xlabel("Labels")
            plt.ylabel("Count")
            plt.xticks(rotation=45)
            plt.savefig("figure2_class_distribution.png")
            plt.show()

        # 3. Heatmap of Missing Values
        plt.figure(figsize=(10, 6))
        sns.heatmap(df.isnull(), cbar=False, cmap='coolwarm', yticklabels=False)
        plt.title("Heatmap: Missing Values")
        plt.savefig("figure3_missing_values_heatmap.png")
        plt.show()

        # 4. Histogram of Numeric Columns
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
        if len(numeric_cols) > 0:
            plt.figure(figsize=(10, 8))
            for col in numeric_cols:
                sns.histplot(df[col], kde=True, bins=15, edgecolor='black', label=col, alpha=0.6)
            plt.legend()
            plt.title("Overlaid Histograms of Numeric Columns")
            plt.xlabel("Value")
            plt.ylabel("Frequency")
            plt.savefig("figure4_overlaid_histograms.png")
            plt.show()

        # 5. Box Plot of Numeric Columns
        if len(numeric_cols) > 0:
            plt.figure(figsize=(12, 6))
            sns.boxplot(data=df[numeric_cols], palette="Set2")
            plt.title("Box Plot of Numeric Columns")
            plt.xlabel("Numeric Columns")
            plt.ylabel("Values")
            plt.savefig("figure5_box_plot.png")
            plt.show()

    except FileNotFoundError:
        print("Error: The file was not found. Please check the CSV path.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# Provide the path to your CSV file
csv_path = r"C:\Users\WALTON\Desktop\final-thesis\dataset\csv\final_dataset.csv"  # Replace with the path to your dataset
generate_figures(csv_path)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
csv_path = r"C:\Users\WALTON\Desktop\final-thesis\dataset\csv\final_dataset.csv"  # Replace with the path to your dataset

# Function to generate figures from CSV
def generate_all_figures(csv_path):
    try:
        # Load the CSV file
        df = pd.read_csv(csv_path)

        # 1. Display the first few rows (Table as a Figure)
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.axis('off')
        pd.plotting.table(ax, df.head(10), loc='center', cellLoc='center', colWidths=[0.2] * len(df.columns))
        plt.title("Table: First 10 Rows of the Dataset")
        plt.savefig("figure1_table_preview.png")
        plt.show()

        # 2. Class Distribution Bar Plot (if 'label' exists)
        if "label" in df.columns:
            plt.figure(figsize=(8, 6))
            df['label'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
            plt.title("Class Distribution")
            plt.xlabel("Labels")
            plt.ylabel("Count")
            plt.xticks(rotation=45)
            plt.savefig("figure2_class_distribution.png")
            plt.show()

        # 3. Heatmap of Missing Values
        plt.figure(figsize=(10, 6))
        sns.heatmap(df.isnull(), cbar=False, cmap='coolwarm', yticklabels=False)
        plt.title("Heatmap: Missing Values")
        plt.savefig("figure3_missing_values_heatmap.png")
        plt.show()

        # 4. Histogram of Numeric Columns
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
        if len(numeric_cols) > 0:
            for col in numeric_cols:
                plt.figure(figsize=(8, 6))
                sns.histplot(df[col], kde=True, bins=15, edgecolor='black')
                plt.title(f"Histogram: {col}")
                plt.xlabel(col)
                plt.ylabel("Frequency")
                plt.savefig(f"figure4_histogram_{col}.png")
                plt.show()

        # 5. Pair Plot of All Numeric Columns
        if len(numeric_cols) > 1:
            sns.pairplot(df[numeric_cols], diag_kind="kde", corner=True)
            plt.suptitle("Pair Plot of Numeric Columns", y=1.02)
            plt.savefig("figure5_pair_plot.png")
            plt.show()

    except FileNotFoundError:
        print("Error: The file was not found. Please check the CSV path.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Provide the path to your CSV file
csv_path = "/path/to/your/csv_file.csv"
generate_all_figures(csv_path)
```
